Your task is to:

Take space separated list of integers as input from the user.
In first line of output print the original list.
In the second line of output print the list after arranging integers values in descending order.


Note:

Refer to the Displayed test cases for a better understanding.
The inputs numbers can be randomly distributed and can be +ve or -ve.


Constraints:

1 <= number of elements <= 100



Sample Test Case:

1 3 2 5 4 -----> Input elements of the list separated by space.

[1, 3, 2, 5, 4]-----> Input original list

[5, 4, 3, 2, 1] ----> Print the list after arranging values in descending order.

```python
# Take input from the user
input_string = input()

# Convert the input string into a list of integers
original_list = list(map(int, input_string.split()))

# Print the original list
print(original_list)

# Sort the list in descending order
sorted_list = sorted(original_list, reverse=True)

# Print the sorted list
print(sorted_list)
```

Your task is to:

Take the elements of the lists in a single line separated by a comma(,) and print the list.
After that, convert each element of the original list that is in string format into an integer, and print the list again.
In the last line of output print the integer divisibility operation (//) where first integer element will be divided by the last integer element in the list.


Note:

 Refer to the Displayed test cases for a better understanding.
The inputs must be numbers (positive or negative integer's).


Constraints:

1 <= number of elements <= 100



Sample Test case:

1,2,3,4,5,6 -----> Input element in the list separated by comma(,).

['1', '2', '3', '4', '5', '6'] -----> Print the original list of elements in which the numbers are in string format.

[1, 2, 3, 4, 5, 6] ----> Print the list of integers.

0--------> Last line print 1 divided by 6 i.e. 0 (floor divisibility operation)

```python
# Take input from the user
input_string = input()

# Convert the input string into a list of strings
original_list = input_string.split(',')

# Print the original list
print(original_list)

# Convert the list of strings to a list of integers
int_list = list(map(int, original_list))

# Print the list of integers
print(int_list)

# Perform floor division of the first integer by the last integer
result = int_list[0] // int_list[-1]

# Print the result of the floor division
print(result)
```

Your task is to:

Take comma (,) separated list of integers as input from the user.
Print the largest as well as the smallest number's available from the list (on single line separated with space)


Note:

 Refer to the Displayed test cases for a better understanding.
The input numbers can be positive or negative integer's.
Data in the list is non unique randomly distributed integer's.


Constraints:

1 <= number of elements <= 100



Sample Test case:

1,4,3,2,5 -----> Input elements of the list separated by comma(,).

5 1 ----> Print the largest and the smallest number's from the list.

```python
# Take input from the user
input_string = input()

# Convert the input string into a list of integers
num_list = list(map(int, input_string.split(',')))

# Find the largest and smallest numbers
largest = max(num_list)
smallest = min(num_list)

# Print the largest and smallest numbers separated by a space
print(largest, smallest)
```

Your task is to:

Take space separated list of integers as input from the user.
In the first line of output, print the original list.
In the second line of output print the list after arranging integer values in descending order.
In the last line of output print the minimum value you find out in the list.


Note:

 Refer to the Displayed test cases for a better understanding.
The input numbers can be randomly distributed and can be +ve or -ve.


Constraints:

1 <= number of elements <= 100



Sample Test Case:

1 3 2 5 4 -----> Input elements of the list separated by space.

[1, 3, 2, 5, 4]-----> Input original list

[5, 4, 3, 2, 1] ----> Print the list after arranging values in descending order.

1------> Minimum value present in the list

```python
# Take input from the user
input_string = input()

# Convert the input string into a list of integers
original_list = list(map(int, input_string.split()))

# Print the original list
print(original_list)

# Sort the list in descending order
sorted_list = sorted(original_list, reverse=True)

# Print the sorted list
print(sorted_list)

# Find and print the minimum value in the original list
min_value = min(original_list)
print(min_value)
```
